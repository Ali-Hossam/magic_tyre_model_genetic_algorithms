# **Algorithm**

### Data Generation

```python
# range of every parameter
xi_min, xi_max = -0.75, 0.75
B_min, B_max = 5, 30
C_min, C_max = 0.5, 2
E_min, E_max = -2, 0
Fz_min, Fz_max = 1, 5
D_min, D_max = 0.1, 2
fx_min, fx_max = -0.02, 0.02
fy_min, fy_max = -0.300, 0.300
dataset_size = 10
input_size = 76
step = 0.02
```

```python
# create array with random values 100k times
B_arr = np.random.uniform(B_min, B_max, dataset_size)
C_arr = np.random.uniform(C_min, C_max, dataset_size)
E_arr = np.random.uniform(E_min, E_max, dataset_size)
Fz_arr = np.random.uniform(Fz_min, Fz_max, dataset_size)
D_arr = np.random.uniform(D_min, D_max, dataset_size)
fx_arr = np.random.uniform(fx_min, fx_max, dataset_size)
fy_arr = np.random.uniform(fy_min, fy_max, dataset_size)
```

```python
print("Size of the dataset is : %2.1fK "%(B_arr.size/1000))
```

```python
# create slip ratio input array
xi_arr = np.arange(-0.75, 0.76, step)
xi_arr.size
```

```python
# magic formula
def MF_func(B_arr, C_arr, E_arr, D_arr, fx_arr, fy_arr, Fz_arr, SR_arr):
    MF_dataset = np.zeros((dataset_size, input_size))
    B = B_arr
    C = C_arr
    E = E_arr
    Fz = Fz_arr
    D = D_arr
    fx = fx_arr
    fy = fy_arr
    
    for i in range(len(SR_arr)):
        xi = SR_arr[i]
        MF_dataset[:, i] = fy+Fz*D*np.sin(C*np.arctan(B*(xi+fx)-E*(B*(xi+fx)-np.arctan(B*(xi+fx)))))
        
    return MF_dataset
```

```python
MF_df = MF_func(B_arr, C_arr, E_arr, D_arr, fx_arr, fy_arr, Fz_arr, xi_arr)
```

```python
for i in range(MF_df.shape[0]):
    plt.plot(xi_arr, MF_df[i])
    plt.xlabel("Slip Ratio")
    plt.ylabel("Fx")
    plt.title("Pure Longitudinal Slip")
```

### Test Algorithm in Pure Longitudinal Slip Mode

```python
from GeneticAlg import GeneticAlgorithm
```

```python
# alg params
mode = "Long"
NP = 50
CP = 0.6
MP = 0.1
F = 0.5
n_params = 14
iter_max = 1000
c = 0
```

```python
def force_pred(b_arr, Fz):
    b0, b1, b2, b3, b4, b5, b6,\
    b7, b8, b9, b10, b11, b12, b13 = b_arr
    Y_pred = []
    for slip_angle in xi_arr:
        D = b1 * pow(Fz, 2) + b2 * Fz
        C = b0
        B = ((b3 * pow(Fz, 2) + b4 * Fz) * np.exp(-b5*Fz))/(C*D)
        Shx = b9 * Fz + b10
        Svx = b11 * Fz + b12
        E = (b6 * pow(Fz, 2) + b7*Fz + b8) * (1 - b13 * np.sign(slip_angle + Shx))

        x = slip_angle
        # y = D * np.sin(np.deg2rad(C * np.arctan(np.deg2rad(B * x - E *
        #                 (B * x - np.arctan(np.deg2rad(B*x))))))) + Svx

        y = D * np.sin(C * np.arctan(B * x  - E * (B*x - np.arctan(B*x)))) + Svx
        y = y / (x+Shx)
        Y_pred.append(y)    # Longitudinal force in pure slip condition
    return Y_pred
```

```python
def plot(i, j, y_true, y_pred, k, axes):
    axes[i][j].plot(xi_arr, y_true, label="True")
    axes[i][j].plot(xi_arr, y_pred, label="Pred")
    axes[i][j].set_title("Curve"+str(k))
    axes[i][j].legend()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(20, 8), sharex=True, sharey=True, constrained_layout=True)
i, j = 0, 0

for k in range(MF_df.shape[0]):
    alg = GeneticAlgorithm(NP, CP, MP, F, mode, xi_arr, MF_df[k], Fz_arr[k], n_params, iter_max, c)
    alg.apply_alg()
    b_arr = alg.x_best
    y_pred = force_pred(b_arr, Fz_arr[k])
    print(alg.min_error)

    # plot
    plot(i, j, MF_df[k], y_pred, k, axes)
    j += 1;
    if j == 5:
        j = 0
        i += 1

fig.supxlabel('Longitudinal Slip Ratio')
fig.supylabel('Fx')
```

### Test algorithm in Pure Lateral Force

```python
# params
a0 = 0.384
a1 = -0.014
a2 = -2.228
a3 = 241.4
a4 = 47.72
a6 = 0.642
a7 = 4.565
a8 = 12.137
a9 = -3.547
a11 = 4.35e-6
a12 = 2.01e-4
a17 = 0.945

a6n = 1e-9
a13n = 1e-9
a16n = 1e-9
a17n = 1e-9
a16 = 1e-9
a3n = 1e-9
a9n = 1e-9

Fz = 8.754
c = 0
```

```python
# force calculations
# c: camber angles
slip_angles = np.arange(-15, 16, 0.5)
lat_forces = []

for slip_angle in slip_angles:
    D = Fz*(a1*Fz + a2) * (1 - a3n * pow(c, 2))
    C = a0
    B = (a3 * np.sin(2*np.arctan(Fz/a4) * (1 - a6n*np.abs(c))))/(C*D)
    Shy = a8*Fz + a9 + a13n*c
    Svy = a11*Fz + a12 + c*(a16n*pow(Fz, 2) + a17n * Fz)
    E = (a6 * Fz + a7)*(1 - (a9n*c + a17 * np.sign(slip_angle + Shy)))

    x = slip_angle
    y = D * np.sin(C * np.arctan(B * x - E *
                                 (B * x - np.arctan(B*x)))) + Svy

    # Longitudinal force in pure slip condition
    Y_pure = y / (x + Shy)
    lat_forces.append(Y_pure)
```

```python
# Visulalize
plt.plot(slip_angles, lat_forces)
plt.xlabel("Side Slip Angle")
plt.ylabel("Fy")
plt.title("Pure Lateral Slip")
```

```python
y_pred = []
for slip_angle in slip_angles:
    D = Fz*(a1*Fz + a2) * (1 - a3 * pow(c, 2))
    C = a0
    B = (a4 * np.sin(2*np.arctan(Fz/a5) * (1 - a6*np.abs(c))))/(C*D)
    Shy = a11*Fz + a12 + a13*c
    Svy = a14*Fz + a15 + c*(a16*pow(Fz, 2) + a17 * Fz)
    E = (a7 * Fz + a8)*(1 - (a9*c + a10 * np.sign(slip_angle + Shy)))

    x = slip_angle
    y = D * np.sin(C * np.arctan(B * x - E *
                             (B * x - np.arctan(B*x)))) + Svy

    # Longitudinal force in pure slip condition
    Y_pure = y / (x + Shy)
    y_pred.append(Y_pure)
```
